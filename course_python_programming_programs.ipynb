**Example**: check the parity of a number.

```python
print('Check the parity of a number.')
n = int(input('Enter an integer : '))

if (n % 2) == 0 :
   print(n, ' is even')
else :
   print(n, ' is odd')
```

**Example**: create three lists containing alphanumeric characters:
* 1st list contains digit characters: codes [48 .. 57]
* 2nd list contains uppercase letter characters: codes [65 .. 90]
* 3rd list contains lowercase letter characters: codes [97 .. 123]

```python
s1, s2, s3 = [], [], []

for k in range(10):
  s1 += [chr(k + 48)]

for k in range(26):
  s2 += [chr(k + 65)]
  s3 += [chr(k + 97)]

print(s1)
print(s2)
print(s3)
```

**Example**: check if a character is alphanumeric.

```python
s1, s2, s3 = [], [], []

for k in range(10):
  s1 += [chr(k + 48)]
for k in range(26):
  s2 += [chr(k + 65)]
  s3 += [chr(k + 97)]

print('Check if a character is alphanumeric.')
c = input('Enter a character : ')

if c in s1 :
  print(c, ' is a digit')
elif c in s2 :
  print(c, ' is an uppercase letter')
elif c in s3 :
  print(c, ' is a lowercase letter')
else :
  print(c, ' is not alphanumeric')
```

**Example**: the Fibonacci sequence.

The Fibonacci sequence is defined as follows:

$F_0 = 1$ and $F_1 = 1$ and $F_n = F_{n-1} + F_{n-2}$ for $n > 1$

```python
k = 10
f0, f1 = 1, 1
print(f0, f1, end=' ')

for _ in range(k) :
  print(f0 + f1, end=' ')
  f0, f1 = f1, f0 + f1
```

**Example**: the 3n + 1 sequence

The rule for creating the sequence 3n + 1 is to start with a positive integer and generate the next term from the previous term, either by dividing it by two when it is even, or by multiplying it by three and adding 1 when it is odd. The sequence ends when the next term reaches 1.

The Collatz conjecture states that the 3n + 1 sequence will always eventually lead to the number one.

```python
n = 7

while n != 1:
    print(n, end=' ')
    if n % 2 == 0:        # n is even
        n = n // 2
    else:                 # n is odd
        n = n * 3 + 1
print(n)                  # the last print is 1
```

**Example**: the Greatest Common Divisor (GCD).

Write a program that calculates the value of the GCD of two natural numbers.

Euclidâ€™s algorithm allows us to find the GCD of two natural numbers $m$ and $n$.

This algorithm works as follows:

* find the remainder of the division of $m$ by $n$,
* then find the remainder of the division of $n$ by the first remainder found,
* and so on until a zero remainder is found.
* The last divisor used is the GDC of $m$ and $n$.

Example

Find the GDC of $m$ = 1386 and $n$ = 140.

|Dividende|Divisor|Euclidian division|Remainder|
|:---|:---|:---|:---|
|1386|140|1386 = 140 x 9 + 126|126|
|140|126|140 = 126 x 1 + 14|14|
|126|**14**|126 = 14 x 9 + 0|**0**|

The last divisor is **14**, and therefore, the GCD of 1386 and 140 is 14.

```python
m, n = 1386, 140

while n != 0 :
  m, n = n, m % n

print(m)
```

**Example**: prime numbers.

The Sieve of Eratosthenes is a process that allows us to find all prime numbers less than a given natural integer N.

The algorithm works by elimination: it involves removing all multiples of an integer (other than itself) from a list of integers from 2 to N.

By removing all these multiples, at the end only the integers that are not multiples of any integer except 1 and themselves will remain, and which are therefore prime numbers.

We start by crossing out the multiples of 2, then the remaining multiples of 3, then the remaining multiples of 5, and so on, each time crossing out all multiples of the smallest remaining integer.

The animation below illustrates the Sieve of Eratosthenes for N=120:

```python
num = 120

prime = []
for i in range(num+1) :
  prime += [True]

p = 2
while (p * p <= num) :
  if (prime[p] == True) :
    for i in range(p * p, num+1, p) :
        prime[i] = False
  p += 1

for p in range(2, num+1) :
  if prime[p] :
    print(p, end=' ')
```
