# Abstraction procédurale

**Procédure**: le résultat de l'abstraction des valeurs réelles utilisées dans un calcul particulier est un modèle de calcul ou une méthode de calcul.

**Décomposition procédurale**: processus de décomposition d'un problème en un certain nombre de sous-problèmes, de sorte que chaque sous-problème accomplisse une tâche identifiable, qui peut elle-même être subdivisée.

**Abstraction procédurale**: simplifier un problème en le décomposant en une série de procédures ou de sous-routines qui sont généralisées avec des paramètres variables.

Cette leçon vise à illustrer l'abstraction procédurale qui implique des concepts tels que la reconnaissance des modèles, l'abstraction et la généralisation, et également illustrer le concept de programmation modulaire.

# Dessin

## Dessiner un carré


```python
home()
clear()
for _ in range(4):
  forward(100)
  right(90)
```

La commande `forward(100)` fait avancer la tortue par `100` unités.

La commande `forward(90)` fait tourner la tortue à droite de `90` degrés.

Les commandes de la tortue ont des équivalents courts:

* `fd()` pour `forward()`
* `rt()` pour `right()`
* `lt()` pour `left()`

# Carrés

Nous écrirrons une fonction qui dessine un carré. Le *côté* du carré est passé comme argument.

```python
def square(side):
  for _ in range(4):
    forward(side)
    right(360 / 4)

home()
clear()

square(50)
square(75)
square(100)
```

# Polygones

Le concept de tour total à 360 ° nous permet de *généraliser* la fonction qui dessine un carré en une fonction qui dessine un polygone régulier, en utilisant le nombre de sommets du polygone en tant que *paramètre*.

```python
def polygon(corners, side):
  for _ in range(corners):
    forward(side)
    right(360 / corners)

home()
clear()

polygon(12, 50)
```

# Fleur de polygone

Une fleur de plolygone est constituée de pétales représentés par des polygones réguliers ayant la même forme qui sont tournés avec le même angle.

```python
def polygon_flower(petals, corners, side):
  for _ in range(petals):
    polygon(corners,side)
    right(360 / petals)

home()
clear()

polygon_flower(5,5,50)
```

# Cercles

Un cercle est approximé par un polygone régulier inscrit.

<br> <br>
course-python-programming-turtle-graphics-03.png
<br> <br>

Nous écrirons une fonction qui dessine un cercle d'un rayon donné, le cercle sera approximé par un polygone régulier avec 360 côtés.

Le périmètre du polygone est une approximation du périmètre du cercle.

Soit $r$ le rayon du cercle et $s$ la longueur du côté polygone, alors nous avons:

$2 \pi r = 360 \times s$

```python
def draw_circle(radius):
  side = 2 * 3.14 * radius / 360
  polygon(360, side)

home()
clear()
speed(13)

draw_circle(50)
draw_circle(75)
draw_circle(100)
```

# Arcs

Une fonction qui dessine un arc de cercle est une généralisation de la fonction qui dessine un cercle, l'angle au centre de l'arc est passé comme paramètre.

```python
def draw_arc(radius, angle):
  side = 2 * 3.14 * radius / 360
  for _ in range(angle):
    forward(side)
    right(1)

home()
clear()

draw_arc(100,120)
```

# Fleur

## Petale

Un pétale peut être considéré comme deux arcs réunis.

<br> <br>
course-python-programming-turtle-graphics-06.png
<br> <br>

Soit $\alpha$ l'angle au centre des arcs composant un pétale.

Après avoir dessiné le premier arc, la tortue doit tourner d'un angle $\theta$ donné par la relation suivante:

$\theta = 180 - \alpha$

Nous écrirons une fonction qui dessine un pétale comme deux arcs réunis.

```python
def petal(radius, angle):
  for _ in range(2) :
    draw_arc(radius, angle)
    right(180 - angle)

home()
clear()

petal(100,120)
```

## Fleur

Nous écrirons une fonction qui dessine une fleur avec un nombre de pétales passé comme paramètre.

```python
def flower(petals, radius, angle):
  for _ in range(petals):
    petal(radius, angle)
    right(360 / petals)

home()
clear()

flower(6,100,90)
```
