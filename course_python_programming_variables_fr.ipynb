# Variables

# Variable

Une variable Python est un nom symbolique qui fait référence à un objet Python.

Une variable est créée automatiquement lorsqu'un objet lui est affecté.

Le signe égal (`=`) est utilisé pour attribuer un objet à une variable.

**Exemple**

```python
n = 300
```

Lorsque l'interpréteur Python exécute cette instruction, il procède ainsi:

* Créer un objet qui est une instance de la classe `int` (entier),
* Attribuer à cet objet la valeur `300`,
* Créer un nom symbolique `n`,
* Déclarer `n` comme référence à cet objet.

# Affichage de la classe d'un objet

La fonction intégrée `type ()` renvoie la classe d'un objet en conséquence.

**Exemple**

```python
n = 256
print(type(n))
```

# Plusieurs références

Un objet n'est créé en mémoire qu'une seule fois et peut être référencé par plusieurs variables.

**Exemple**

```python
m = n
```

Lorsque l'interpréteur Python exécute cette instruction, elle ne crée pas un nouvel objet.

Il crée simplement un nouveau nom symbolique `m` qui fait référence au même objet référencé par `n`.

# Changement de référence

**Exemple**

```python
m = 400
```

Lorsque l'interpréteur Python exécute cette instruction, il crée un objet qui est une instance de la classe `int`, attribue la valeur `400` à cet objet et déclare `m` comme référence à cet objet.

<br> <br>
course-python-programming-variables-03.png
<br> <br>

**Exemple**

```python
n = 'foo'
print(type(n))
```

Lorsque l'interpréteur Python exécute cette instruction, il crée un objet qui est une instance de la classe `str` (string), attribue la valeur `foo`à cet objet et déclare `n` comme référence à cet objet.

<br> <br>

L'objet de la classe `int` ayant la valeur `300` et précédemment référencé par `n` n'est plus référencé par aucun nom et n'est plus accessible, il est devenu orphelin.

Un processus appelé *garbage collector* nettoiera cet objet à partir de la mémoire et libérera l'emplacement de la mémoire qui lui est allouée

# Afficher

La fonction prédéfinie `print()` est utilisée pour afficher la valeur d'une variable.

**Exemple**

```python
counter = 1
print(counter)
```

# Supprimer

L'instruction `del` est utilisée pour supprimer une variable.

**Exemple**

```python
del counter
```

# Casting

Le casting permet de spécifier la classe d'un objet référencé par une variable.

**Exemple**

```python
x = str(10)
y = int(10)
z = float(10)
```

`x` référence à un objet de la classe `str`: sa valeur sera la chaîne de caractères `'10'`.

`y` référence à un objet de la classe `int`: sa valeur sera le nombre entier `10`.

`z` référence à un objet de la classe `float`: sa valeur sera le nombre réel `10.0`.

# Identité d'un objet

Chaque objet créé se voit attribuer un nombre qui l'identifie de manière unique.

La fonction prédéfinie `id()` renvoie comme résultat un nombre entier identifiant un objet.

**Exemple**

```python
n = 300
print(id(n))
m = 400
print(id(m))
```
