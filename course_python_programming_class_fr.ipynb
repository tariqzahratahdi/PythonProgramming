# Objet

Le modèle de programmation orientée objet (OOP) organise la conception des logiciels autour des données appelées **objets**.

Un objet est l'unité de base de la programmation orientée objet.

Il permet de modéliser un concept ou une entité du monde physique.

Il est décrit par une *identité*, un *état* et un *comportement*.

**Exemple**: un objet modélisant une voiture

Un objet est une structure de données qui répond à un ensemble de messages.

Cette structure de données définit son état tandis que l'ensemble de messages qu'il inclut décrit son comportement:

Les données, ou champs, qui décrivent sa structure interne sont appelées ses **attributs**.

La réponse à la réception d'un message par un objet est appelée une **méthode** (méthode d'implémentation du message); Elle décrit quelle réponse doit être donnée au message.

# Classe

L'objectif principal de la programmation orientée objet est de définir les données et les fonctions qui fonctionnent dessus dans une seule entité appelée **classe**, de sorte qu'aucune autre partie du programme ne peut accéder à ces données, à l'exception de ces fonctions.

Une classe est un type de données défini par l'utilisateur.

Un objet est une *instance* d'une classe.

Une classe est un modèle (plan, prototype) utilisé pour créer un objet et définit les *attributs* et les *méthodes* associés à cet objet.

Exemple: modélisation de classe Une voiture

# Classes et objets Python

Un objet est une *encapsulation* de variables (*attributs*) et de fonctions (*méthodes*) en une seule entité.

Un objet obtient ses attributs et ses méthodes à partir d'une classe.

Un objet est une *instance* d'une classe.

**Exemple**: Classe Modélisation d'une voiture



```python
# Définir une classe Car
class Car:
    company = ""
    brand = ""
    engine = ""
    color = ""

    def description(self):
        print(f'{self.company} {self.brand} {self.engine} {self.color}')

# créer un objet à partir de la classe Car
my_car = Car()

# accéder aux variables de l'objet
my_car.company ="Toyota"
my_car.brand = "Yaris"
my_car.engine = "Diesel"
my_car.color = "Blue"

# Accédez aux fonctions de l'objet
my_car.description()
```

# **`self`**

**`self`** est le premier argument de chaque méthode d'une classe Python.

Lorsqu'on appelle une méthode sur une instance d'une classe, le premier argument transmis à cette méthode est cette instance.

Dans Python, nous utilisons le mot `self` pour représenter cette instance.

L'utilisation du mot `self` n'est qu'une convention Python.

Certains langages de programmation utilisent le mot `this` pour représenter cette instance.

# **`init()`**

La fonction **`__init__()`** est une fonction spéciale qui est appelée lorsque la classe est initiée. Elle est utilisée pour attribuer des valeurs dans une classe.

**Exemple**

```python
# Définir une classe de Car
class Car:
    company = ""
    brand = ""
    engine = ""
    color = ""

    def __init__(self, com, bra, eng, col):
        self.company = com
        self.brand = bra
        self.engine = eng
        self.color = col

    def description(self):
        print(f'{self.company} {self.brand} {self.engine} {self.color}')

# créer un objet à partir de la classe Car
my_car = Car("Toyota", "Yaris", "Gasoline", "Blue")

# Accédez aux fonctions de l'objet
my_car.description()
```
