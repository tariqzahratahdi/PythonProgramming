# Procedural Abstraction

**Procedure**: the result of abstracting away the actual values used in any particular computation is a computational pattern or computational method.

**Procedural decomposition**: the process of breaking down a problem into a number of sub-problems, so that each sub-problem accomplishes an identifiable task, which might itself be further subdivided.

**Procedural abstraction**: simplifying a problem by breaking it down into a series of procedures or subroutines that are generalized with variable parameters.

This lesson aims at illustrating the procedural abstraction wich involves concepts such as pattern recognition, abstraction, generalization, and also to illustrate the modular programming concept.

# Drawing

## Draw a square


```python
home()
clear()
for _ in range(4):
  forward(100)
  right(90)
```

The command `forward(100)` moves the turtle forward by `100` units.

The command `right(90)` turns the turtle to the right by `90` degrees.

Turtle commands have short equivalents:

* `fd()` for `forward()`
* `rt()` for `right()`
* `lt()` for `left()`

# Squares

We will write a function that draws a square. The *side* of the square is passed as a parameter.

```python
def square(side):
  for _ in range(4):
    forward(side)
    right(360 / 4)

home()
clear()

square(50)
```

# Polygons

The 360° total turn concept allows us to *generalize* the function that draws a square to a function that draws a regular polygon, using the number of *corners* of the polygon as a *parameter*.

```python
def polygon(corners, side):
  for _ in range(corners):
    forward(side)
    right(360 / corners)

home()
clear()

polygon(6, 100)
```

# Polygon Flower

A plolygon flower has petals made of polygons with the same shape that are rotated with the same angle.

```python
def polygon_flower(petals, corners, side):
  for _ in range(petals):
    polygon(corners,side)
    right(360 / petals)

home()
clear()

polygon_flower(5,5,50)
```

# Circles

A circle is approximated by an inscribed regular polygon.

<br><br>
course-python-programming-turtle-graphics-03.png
<br><br>

We will write a function that draws a circle of a given radius, the circle will be approximated by a regular polygon with 360 sides.

The perimeter of the polygon is an approximation of the perimeter of the circle.

Let $r$ be the radius of the circle and $s$ be the length of the polygon side, then we have:

$2 \pi r = 360 \times s$

```python
def draw_circle(radius):
  side = 2 * 3.14 * radius / 360
  polygon(360, side)

home()
clear()
speed(13)

draw_circle(50)
```

# Arcs

A function that draws an arc of a circle is a generalization of the function that draws a circle, the central angle of the arc is passed as a parameter.

```python
def draw_arc(radius, angle):
  side = 2 * 3.14 * radius / 360
  for _ in range(angle):
    forward(side)
    right(1)

home()
clear()

draw_arc(100,90)
```

# Flower

A flower is made of petals.

<br><br>

## Petal

A petal can be viewed as two arcs that are joined together.

<br><br>
course-python-programming-turtle-graphics-06.png
<br><br>

Let $\alpha$ be the central angle of the arcs composing a petal.

After drawing the first arc, the turtle will turn by an angle $\theta$ given by the following relation:

$\theta = 180 - \alpha$

We will write a function that draws a petal as two arcs joined together.

```python
def petal(radius, angle):
  for _ in range(2) :
    draw_arc(radius, angle)
    right(180 - angle)

home()
clear()

petal(100,90)
```

## Flower

We will write a function that draws a flower with a number of petals passed as a parameter.

```python
def flower(petals, radius, angle):
  for _ in range(petals):
    petal(radius, angle)
    right(360 / petals)

home()
clear()

flower(4,100,90)
```
